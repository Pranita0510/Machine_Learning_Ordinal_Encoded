```python
import pandas as pd
```

```python
#sample data
data={
    'Customer_id':['C001','C002','C003','C004','C005','C006','C007','C008'],
    'Education_level':['High school','Bachelors','Masters','phd','Bachelors','High school','phd','Masters'],
    'Product_Quality':['Low','Medium','High','Medium','Low','High','Medium','High'],
    'Satisfaction_Level':['Poor','Average','Excellent','Good','Average','Poor','Good','Excellent']
}
df=pd.DataFrame(data)
df
```

```python
#Define the order for each ordinal feature
education_order=['High school','Bachelors','Masters','phd']
Product_quality_order=['Low','Medium','High']
Satisfaction_order=['Poor','Average','Good','Excellent']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder=OrdinalEncoder(categories=[education_order,Product_quality_order,Satisfaction_order])
```

```python
#apply encoding(skip Customerid as it is identification)

df[['Education_level','Product_Quality','Satisfaction_Level']]=encoder.fit_transform(df[['Education_level','Product_Quality','Satisfaction_Level']])
df
```
